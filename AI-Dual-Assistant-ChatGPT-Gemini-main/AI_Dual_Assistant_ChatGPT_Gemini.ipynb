```python
# Import and Configure APIs
from openai import OpenAI
import google.generativeai as genai
```

```python
# üîë Replace with your actual keys
openai_key = "YOUR_OWN_API_KEY"    # ChatGPT Key
gemini_key = "YPUR_OWN_API_KEY"  # Gemini Key
```

```python
# Configure Clients
client = OpenAI(api_key=openai_key)
genai.configure(api_key=gemini_key)
```

```python
# Create Gemini Model
gemini_model = genai.GenerativeModel("gemini-2.5-pro")

# -- Functions --

# ChatGPT Function
def ask_chatgpt(question):
    try:
        response = client.chat.completions.create(
            model="gpt-4.1-mini",       # Correct model
            messages=[{"role": "user", "content": question}]
        )
        return response.choices[0].message.content

    except Exception as e:
        return f"‚ùå ChatGPT Error: {str(e)}"


# Gemini Function
def ask_gemini(question):
    try:
        response = gemini_model.generate_content(question)
        return response.text

    except Exception as e:
        return f"‚ùå Gemini Error: {str(e)}"



# Main Program

question = input("Enter your question: ")

print("\n==============================")
print("üîµ ChatGPT Response")
print("==============================")
print(ask_chatgpt(question))

print("\n==============================")
print("üü¢ Gemini Response")
print("==============================")
print(ask_gemini(question))
```
